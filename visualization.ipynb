# 1. Barplot


What it is used for:

A barplot compares numerical values across categories.

Use it when:

X-axis = category (Country, Gender, Product, etc.)
              Vs
Y-axis = number (Salary, Population, Score, etc.)

Example questions barplot answers:

Which country has the highest population?

Which product has the highest sales?

What is the average salary for each job type?

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample data
df = pd.read_csv("demo.csv")

plt.figure(figsize=(8,5))

sns.barplot(data=df, x="Country", y="Population")
plt.xlabel("Country")
plt.ylabel("Population")
plt.title("Population by Country")
# plt.ticklabel_format(style='plain', axis='y')  # <-- disable scientific notation
plt.show()
```

### 2. Countplot


What it is used for:

A countplot shows how many times each category appears in the dataset.

Use it when:

You want to know frequency or counts of categories.

You do NOT need a numeric column.

Example questions countplot answers:

How many male vs female?

How many customers are from each country?

How many transactions occurred per day?

```python
sns.countplot(data=df_frame,x="Country")
```

### 3 Scatter Plot (Relationship between 2 variables)

What it is used for:

A scatter plot shows the relationship between two numerical variables.

Use it when:

X-axis = number

Y-axis = number

Example questions scatter plot answers:

Does house size affect house price?

Does study time affect exam score?

Does temperature affect electricity usage?

```python
sns.scatterplot(data=df_price, x="HouseSize_sqft", y="Price_USD")
plt.title("House Size vs Price")
plt.show()
```

### 4 Histogram (Distribution of a variable)

A histogram helps answer:

✔ “How is my numeric variable distributed?”
✔ “Where do most values fall?”
✔ “Are there outliers?”
✔ “Is the data normal or skewed?”

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(data=df_price, x="Price_USD", bins=10, kde=True)
plt.title("House Price Distribution")
plt.show()
```

### 5 Boxplot (Detect Outliers)

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame({
    # "Right_Skewed": [5, 7, 8, 9, 50],
    "Left_Skewed": [2, 40, 42, 45, 47],
    # "Normal": [10, 12, 15, 18, 20]
})

# -------- BOX PLOT --------
plt.figure(figsize=(6, 6))
# sns.boxplot(data=data)
sns.boxplot(x="Left_Skewed", data=data)

plt.title("Boxplots: Right Skew | Left Skew | Normal")
plt.show()
sns.histplot(data=data,  bins=10, kde=True)
plt.title("House Price Distribution")
plt.show()
```

### Line Plot (Time series)

```python
import matplotlib.pyplot as plt
import pandas as pd

data = {
    "month": ["Jan","Feb","Mar","Apr","May"],
    "sales": [200, 250, 300, 280, 350]
}

df = pd.DataFrame(data)

plt.plot(df["month"], df["sales"], marker="o")
plt.title("Sales Over Time")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```
