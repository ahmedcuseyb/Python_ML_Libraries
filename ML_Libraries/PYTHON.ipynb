# Lists

## Lists store multiple values and are mutable (changeable).

```python
scores = [45, 67, 89, 90]
scores
```

## accessing Elements Using their Index

```python
scores[0]
scores[-1]
```

# Modify List

```python
scores = [45, 67, 89, 90]

# Add a value
scores.append(95)  
# Now: [45, 67, 89, 90, 95]

# Modify a value
scores[1] = 70  
# Now: [45, 70, 89, 90, 95]

# Remove a value
scores.remove(90)
# Now: [45, 70, 89, 95]

print(scores)
```

# Tuples
Tuples are immutable lists — values cannot be changed.

Useful when storing constant values (e.g., coordinates, model setting ranges).

```python
coordinates = (10, 20)
sizes = ("Small", "Medium", "Large")
sizes
```

# Dictionaries
Dictionaries store data in key–value format.
Very useful in encoding, mapping, and configuration.

```python
student = {
    # key and value
    "name": "Ali",
    "age": 21,
    "grade": "A"
}
```

## Access values:

```python
student["name"]
```

## Add or update:

```python
student["age"] = 22
student["city"] = "Nairobi"
```
