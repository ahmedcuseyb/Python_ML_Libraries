```python
import numpy as np

print("\n1. CREATING ARRAYS")
# 1D array
a = np.array([1, 2, 3])
print("a =", a)
# 2D array
b = np.array([(1.5, 2, 3), (4, 5, 6)], dtype=float)
print("\nb =\n", b)
# 3D array
c = np.array([[(1.5,2,3), (4,5,6)],
              [(3,2,1), (4,5,6)]], dtype=float)
print("\nc =\n", c)
# functions
print("\nnp.zeros((3,4)) =\n", np.zeros((3,4)))
print("\nnp.ones((2,3,4)) =\n", np.ones((2,3,4), dtype=np.int16))
print("\nnp.arange(10,25,5) =", np.arange(10,25,5))
print("\nnp.linspace(0,2,9) =", np.linspace(0,2,9))

print("\nnp.full((2,2),7) =\n", np.full((2,2),7))
print("\nnp.eye(2) =\n", np.eye(2))
print("\nnp.random.random((2,2)) =\n", np.random.random((2,2)))
print("\nnp.empty((3,2)) =\n", np.empty((3,2)))
print("seed",np.random.seed(2))
```

### Indexing & Slicing

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([(1.5,2,3),(4,5,6)], dtype=float)
c = np.array([[(1.5,2,3),(4,5,6)],
              [(3,2,1),(4,5,6)]], dtype=float)

print("3. INDEXING & SLICING")

print("a[2] =", a[2])
print("b[1,2] =", b[1,2])
print("a[0:2] =", a[0:2])

print("Reversed a:", a[::-1])

print("a[a < 2] =", a[a<2])

print("b[[1,0,1,0],[0,1,2,0]] =", b[[1,0,1,0],[0,1,2,0]])


```

```python
a = np.array([1, 2, 3, 4, 5])

print(a[1:4])   # [2, 3, 4]
print(a[:3])    # [1, 2, 3]
print(a[3:])    # [4, 5]
print(a[-2:])   # last 2 → [4, 5]
print(a[::-1])  # reverse → [5, 4, 3, 2, 1]
```

```python
# slicing
test = np.array([[1,2,3],[4,5,6]])

print(test[0,0])    # 1
print(test[:,1])   # [2,5]
print(test[1,:])    # [4,5,6]
```

### Inspecting Arrays

```python
import numpy as np

a = np.array([1,2,3])
b = np.array([(1.5,2,3),(4,5,6)], dtype=float)

print("5. INSPECTING ARRAYS")

print("Shape of a:", a.shape)
print("Length of a:", len(a))
print("Dimensions of b:", b.ndim)
print("Size of b:", a.size)
print("Size of b:", b.size)
print("dtype of b:", b.dtype)
print("dtype name:", b.dtype.name)
print("b as int:\n", b.astype(int))
```

### Array Mathematics

```python
import numpy as np

a = np.array([[1,2,3],[4,5,6]])
b = np.array([[1,2,3],[4,5,6]])

print("6. ARRAY MATH")

print("a + b =\n", np.add(a,b))
print("a - b =\n", np.subtract(a,b))
print("a * b =\n", np.multiply(a,b))
print("a / b =\n", np.divide(a,b))

print("\nElement-wise sin(a):\n", np.sin(a))
print("Element-wise sqrt(b):\n", np.sqrt(b))
print("Dot product:\n", a.dot(b.T))
```

### Comparison

```python
import numpy as np

a = np.array([[1,2,3],[4,5,6]])
b = np.array([[1,2,3],[4,5,6]])

print("7. COMPARISON")

print("a == b:\n", a==b)
print("a < 2:\n", a<2)
print("Array equal:", np.array_equal(a,b))
```

### Aggregates

```python
import numpy as np

a = np.array([[1,2,3],[4,5,6]])
b = np.array([[1,2,3],[4,5,6]])

print("8. AGGREGATES")

print("Sum:", a.sum())
print("Min:", a.min())
print("Max (axis=0):", b.max(axis=0))
print("Cumulative sum (axis=1):\n", b.cumsum(axis=1))
print("Mean:", a.mean())
print("Median:", np.median(b))
print("Std dev:", np.std(b))
```
