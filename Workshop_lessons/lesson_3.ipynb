### 1- Manual Encoding Mapping

```python
# Clean strings

scaled_data["Neighborhood"] = scaled_data["Neighborhood"].astype(str).str.strip()
scaled_data["Condition"] = scaled_data["Condition"].astype(str).str.strip()

# Manual mapping dictionaries
neighborhood_map = {
    "Low-Cost": 0,
    "Outskirts": 1,
    "Middle Class": 2,
    "Expensive Area": 3
}

condition_map = {
    "Poor": 0,
    "Fair": 1,
    "Good": 2,
    "Excellent": 3
}

# Apply mapping
scaled_data["Neighborhood"] = scaled_data["Neighborhood"].map(neighborhood_map)
scaled_data["Condition"] = scaled_data["Condition"].map(condition_map)
```
