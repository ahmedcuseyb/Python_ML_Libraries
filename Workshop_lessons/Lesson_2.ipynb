# Methods of Handling Outliers

### 1-StatisticalMethods
### 2. Zcore,
### 3. IQR method,
### 4-Machine Learning (Isolation Forest),

```python
from scipy import stats
zscore=np.abs(stats.zscore(clipping[no_cols]))
solv=(clipping[no_cols])[(zscore>3).any(axis=1)]
# solv=(df_SimpleImputer[no_cols])[(zscore>3).any(axis=1)]

solv
```

```python
solved=solv.reset_index(drop=True)
solved
```

```python
outlier_free = clipping[(zscore > -3).all(axis=1) & (zscore < 3).all(axis=1)]
outlier_free
# <-3, >3- Non outlier
```

```python
outlier_free=outlier_free.reset_index(drop=True)
outlier_free
```

```python
zscore[:5]
```

```python
quartile=clipping[no_cols]

q1=quartile.quantile(0.25)
q3=quartile.quantile(0.75)
iqr=q3-q1
lower_b=q1-1.5*iqr
upper_b=q3+1.5*iqr
# Create a Boolean mask for outliers
outliers_quartile=(quartile < lower_b) | (quartile > upper_b)
# filter outliers and non utliers
outliers_iqr = clipping[outliers_quartile.any(axis=1)]
# outliers_iqr
```

```python
quartile
```

```python
outliers_quartile
```

```python
outliers_iqr=outliers_iqr.reset_index(drop=True)
outliers_iqr
outliers_iqr
```

```python
cleaned_outlier = clipping[~outliers_quartile.any(axis=1)]

cleaned_outlier
```

# Using ML Algo

```python
### IsolationForest
```

```python
from sklearn.ensemble import IsolationForest
import pandas as pd

# Step 1: Select numeric columns
removed_iso = clipping.copy()

X = clipping[no_cols]

# Step 2: Create and fit Isolation Forest
# Parameters
n_estimators = 100  # Number of trees
contamination = 0.05  # Expected proportion of anomalies
sample_size = 256  # Number of samples used to train each tree
iso = IsolationForest(n_estimators=n_estimators,
                            contamination=contamination,
                            max_samples=sample_size,
                            random_state=42)
iso.fit(X)

# Step 3: Predict and label outliers
outlierss = iso.predict(X)
removed_iso['outlier'] =outlierss
# Step 4: Separate  and outliers and non-outliers
Cleaned_iso= removed_iso[removed_iso['outlier'] == 1]
outliers = removed_iso[removed_iso['outlier'] == -1]

# Step 5: Display result
print(f"Detected {len(outliers)} outliers out of {len(removed_iso)} total rows.")
outliers.head()
```
